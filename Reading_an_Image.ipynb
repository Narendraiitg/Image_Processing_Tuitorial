***1) Using Pillow library to load an image***

```python
from PIL import Image

img=Image.open('/content/butterfly.png')
```

```python
img.show()
print(img.format)
```

We need to note that when we load an image using above method than image will not be in the form of numpy array rather it will be in the same format ie JPG/PNG so we can not do any image processing on the loaded image before converting them into numpy array

```python
import numpy as np
```

```python
# Converting an image (JPG/PNG) loaded by pillow library into the numpy array
img1= np.asarray(img)
```

***2) Using Matplotlib library to load an image***

```python
import matplotlib.image as mpimg # Library to load the image 
import matplotlib.pyplot as plt # Library to visualize an image
```

```python
img= mpimg.imread('/content/butterfly.png')
```

Image loaded here will be automatically in numpy array ie no need to convert from JPG/PNG to numpy form

```python
print(type(img)) # Checking the data type of an image loaded using matplotlib.image
```

```python
print(img.shape) # Checking the image dimensions 
```

```python
plt.imshow(img) # Displaying an image using matplotlib.pyplot
plt.colorbar() # Adding the color bar with an image
```

***3) Using sciket library to load an image***

Sciket is also an image processing library in python which can we used for image segmentation, Geometric transformation, color space manupulation,analysis, image filtering, feature extracion/detection.

Also it is used in variousd ML models such as SVM, Random forest etc

```python
from skimage import io # Libaray to read an image
```

```python
sci_img= io.imread('/content/butterfly.png')
```

Unlike an matplot library loading an image from sciket is automatically converted to a numpy array

```python
print(type(sci_img)) #cheking the data type of image loaded using sciket
```

```python
plt.imshow(sci_img)
```

One more aditional benifit using sciket is that we can convert the image data type into another form like float or unsigned interger the same is shown below 

```python
from skimage import img_as_float, img_as_ubyte
```

```python
sci_img_float=img_as_float(sci_img) # converting the arrays values in to a float that is the values between 0-1 
```

```python
print(type(sci_img_float))
```

Above float image is also a numpy array but the values of the arrays are between 0-1 which is simpy converted by dividing each values by the maximum value of the pixel exist in an image mostly divide each value by 255

***4) Using Open CV library to load an image***

Library of programming functions that is mainly aimed at computer vision ie it is useful for Image/video/including live video...

Some of the applications are image classification, object decision, Motion tarcking, Optical Character Recognition OCR etc.

```python
import cv2
```

```python

#Reading an image using open CV
img_cv= cv2.imread('/content/butterfly.png')
```

```python
plt.imshow(img_cv)
```

We can notice that the color is changed automatically, becoze CV2 load the image in the BGR format instead of RGB

Still we can convert the image from BGR to RGB easily

```python
plt.imshow(cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB))
```
