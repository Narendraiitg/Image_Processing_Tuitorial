```python
# Importing the usefull libraries
from skimage import  io
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Reading an image
my_image=io.imread("/content/butterfly.png")
```

```python
# Cheking the data type of an image
type(my_image)
# Can also print the numpy array of image using print(my_image)
```

```python
# Checking the shape of an image
my_image.shape
```

```python
# Displaying the image
plt.imshow(my_image)
print(my_image.min(), my_image.max())
```

From above we can see that the maximum value of the pixel in the input image is 252  where as min value of the pixel is 0

```python
# creating the random image
ran_img=np.random.random([256,256])
plt.imshow(ran_img)
print(ran_img.min(), ran_img.max())
```

From above we can see that the maximum value of the pixel in randomly genarated image is 0.99  where as min value of the pixel is 2.8020479724055747e-05

```python
# Converting the pixels values in the floating from between 0 & 1

from skimage import img_as_float

my_float_img=img_as_float(my_image)

plt.imshow(my_float_img)
print(my_float_img.min(), my_float_img.max())
```

So we have successfully converted the pixels value from int form ie between 0-252 to floating form ie 0-1

```python
# We can aslo replace some of the pixels of our image with some other values 
my_image[10:50,10:50, :]=[255,0,0]
```

```python
plt.imshow(my_image)
```

We just replaced the pixels values between 10-50 with 255 ie red color
